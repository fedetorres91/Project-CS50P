```python
from database import db
from services import Wallet
import datetime
```

```python
# take last_name, email (required), optional first_name, currency)
user = {"last_name":None, "email":None}
while user["last_name"] is None:
    user["last_name"] = input("Last name:")
while user["email"] is None:
    user["email"] = input("email")
db.execute("INSERT INTO users (last_name, email) VALUES (?, ?)", user["last_name"], user["email"])
```

```python
# create wallet
federico = Wallet(1, 1000)
db.execute("INSERT INTO wallet (user_id, balance) VALUES (?, ?)", federico.user_id, federico.balance)
```

```python
# add tranasactions
print(federico.categories)
```

```python
# income
federico.transaction("Income", 500)
db.execute(
    "INSERT INTO transactions (user_id, type, date, amount)" \
    " VALUES (?, ?, ?, ?)", 
    federico.user_id,
    "Income",
    datetime.date.today().isoformat(),
    500)
# update wallet
db.execute("UPDATE wallet " \
"SET balance= ?" \
" WHERE user_id = ?",
federico.balance,
federico.user_id)
```
